##### PaddleOCR = Intantiate and detect our model
##### draw_ocr = Used for visualization purpose

```python
# languagle we are using is english
ocr = PaddleOCR(use_angle_cls=True, lang='en')
```

```python
roi_img_path = '/content/drive/MyDrive/Automatic Number Plate Reader/notebooks/Test/region_of_intrest.jpeg'
result = ocr.ocr(roi_img_path, cls=True)
```

```python
result
```

```python
for idx in range(len(result)):
        res = result[idx]
        for line in res:
            print(line)
result = result[0]
```

```python
from PIL import Image
import os,shutil
```

```python
file_name = '/content/drive/MyDrive/Automatic Number Plate Reader/notebooks/ocr_result.jpeg'
```

```python
image = Image.open(roi_img_path).convert('RGB')
```

```python
boxes = [line[0] for line in result]
txts = [line[1][0] for line in result]
scores = [line[1][1] for line in result]
font_path = os.path.join('/content/drive/MyDrive/Automatic Number Plate Reader/notebooks', 'simfang.ttf')
im_show = draw_ocr(image, boxes, txts, scores, font_path=font_path)
im_show = Image.fromarray(im_show)
ocr_result_img_path = os.path.join('/content/drive/MyDrive/Automatic Number Plate Reader/notebooks', 'predictions', 'ocr_result')
os.makedirs(ocr_result_img_path, exist_ok=True)
ocr_result_img_path = os.path.join('/content/drive/MyDrive/Automatic Number Plate Reader/notebooks', 'predictions', 'ocr_result', file_name)
im_show.save(ocr_result_img_path)
```
